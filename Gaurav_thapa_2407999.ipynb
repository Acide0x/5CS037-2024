```python
import pandas as pd
df=pd.read_csv('/content/gdrive/My Drive/Datasets/Titanic-Dataset.csv')
```

```python
# Extract   'Fare' column and cmake a new DataFrame
fare = df[["Fare"]]

# Display the first 5 rows of the 'fare' DataFrame
fare.head()
```

```python
# Select the 'Pclass' and 'Age' columns to make a new DataFrame
class_age = df[["Pclass", "Age"]]

# Show the first 5 rows of the 'class_age' DataFrame
class_age.head()
```

```python
# make a DataFrame with the 'Survived' and 'Sex' columns
survived_gender = df[["Survived", "Sex"]]

# show the first 5 rows of the 'survived_gender' DataFrame
survived_gender.head()
```

```python
# Select passengers with a 'Fare' greater than 100 and create a new DataFrame
fare_gt_100 = df[df["Fare"] > 100]

# Print the resulting DataFrame
fare_gt_100.head()
```

```python
# Filter passengers in first class (Pclass = 1)
first_class = df[df["Pclass"] == 1]

# Print the result
first_class.head()

```

```python
# put passengers who are female and under 18
female_under_18 = df[(df["Age"] < 18) & (df["Sex"] == "female")]

# Print the result
female_under_18.head()
```

```python
# put passengers whose Embarked port is either 'C' or 'S'
embarked_c_or_s = df[df["Embarked"].isin(["C", "S"])]

# Print the result
embarked_c_or_s.head()
```

```python
# Filter passengers in first or second class
first_second_class = df[df["Pclass"].isin([1, 2])]

# Print the result
first_second_class.head
```

```python
#put  a column 'fare per year' by divide Fare by Age
df["fare per year"] = df["Fare"] / df["Age"]

# Print the result
df.head()
```

```python
# add rows where 'fare per year' is higher than 5
high_fare_age = df[df["fare per year"] > 5]

# Print the result
high_fare_age.head()
```

```python
# Sort 'high fare age' by 'fare per year' in descending order
high_fare_age_srt = high_fare_age.sort_values(by="fare per year", ascending=False)

# Display the sorted result
high_fare_age_srt.head()
```

```python
# put 'Name' and 'fare per year' columns from 'high fare age srt'
result = high_fare_age_srt[["Name", "fare per year"]]

# show the first 5 rows of the result
result.head()
```

```python
# Put a column 'fare per class' and divide Fare by Pclass
df["fare per class"] = df["Fare"] / df["Pclass"]

# Show the first 5 rows to check the result
df.head()
```

```python
# Add rows where the passenger is male and an adult
adult_males = df[(df["Sex"] == "male") & (df["Age"] >= 18)]

# Show  the first 5 rows to check the result
adult_males.head()
```

```python
# Sort 'adult males' by 'fare per class' in descending order
adult_males_srt = adult_males.sort_values(by="fare per class", ascending=False)

# show  the first 5 rows to check the result
adult_males_srt.head()
```

```python
# Choose  'Name', 'Age', and 'fare per class' columns from 'adult males srt'
result = adult_males_srt[["Name", "Age", "fare per class"]]

# Show  the first 5 rows to check the result
result.head()
```

```python
# Show  total Fare paid by all passengers
total_fare = df["Fare"].sum()

# Show the result
total_fare
```

```python
# Put passengers in first class (Pclass is 1)
first_class_passengers = df[df["Pclass"] == 1]

# Show the total Fare paid by first class passengers
total_fare_first_class = first_class_passengers["Fare"].sum()

# printing the result
total_fare_first_class
```

```python
# Show passengers riding in second class (Pclass is 2)
second_class_passengers = df[df["Pclass"] == 2]

# Calculate  total fare paid by the second-class travelers
total_fare_second_class = second_class_passengers["Fare"].sum()

# Show the result for second class
total_fare_second_class

# EXTRACT the passengers riding in third class (Pclass is 3)
third_class_passengers = df[df["Pclass"] == 3]

# Calculate the total fare paid by the third-class travelers
total_fare_third_class = third_class_passengers["Fare"].sum()

# Printing the result for third class
total_fare_third_class
```

```python
# Calculate the total fare for first, second, and third class passengers
total_fare_first_class = df[df["Pclass"] == 1]["Fare"].sum()
total_fare_second_class = df[df["Pclass"] == 2]["Fare"].sum()
total_fare_third_class = df[df["Pclass"] == 3]["Fare"].sum()

# Mix the totals into a list
fare_totals = [total_fare_first_class, total_fare_second_class, total_fare_third_class]

# Printing the fare totals list
fare_totals
```

```python
# Total fare for all passengers
total_fare = df["Fare"].sum()

# Calculate the proportion of fare revenue for each class
proportion_first_class = total_fare_first_class / total_fare
proportion_second_class = total_fare_second_class / total_fare
proportion_third_class = total_fare_third_class / total_fare

# Mixing the proportions into a list
proportions = [proportion_first_class, proportion_second_class, proportion_third_class]

# Print the proportions list
proportions
```

```python
# Show a new column 'age group' to categorize passengers based on age
df["age group"] = pd.cut(df["Age"], bins=[0, 17, 64, float('inf')], labels=["child", "adult", "senior"])

# Displpay the first 5 rows to check the result
df.head()
```

```python
# Calculate the total number of passengers on the Titanic
total_passengers = len(df)

# Show the result
total_passengers
```

```python
# Count the number of passengers in each age group
age_group_count = df["age group"].value_counts()

# Show the result
age_group_count
```

```python
# Calculate proportion of passengers in every  age group
age_group_proportion = age_group_count / total_passengers

# Show  the result
age_group_proportion
```
